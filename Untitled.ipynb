## Q17). Total runs across each season.

## Q20). Top 10 Teams who scores Maximum runs in first 5 overs

## Q24). Chances of chasing 200+ target

```python
target_200 = ds2.groupby(['match_id', 'inning', 'batting_team' ])['total_runs'].sum().reset_index()
runs = target_200['total_runs'].values
runs_1st = runs[::2]
runs_2nd = runs[1::2]
win_chase = 0
runs_200 = 0
for i,j in zip(runs_1st, runs_2nd):
    if i>=200:
        runs_200+=1
        if j>i:
            win_chase+=1
    
print(f'The number of times 200+ runs were made:-\t {runs_200}')
print(f'The number of times 200+ target were chased:-\t {win_chase}')
```

## Q25). Top 10 Batsman with most number of Fours.

```python
batsman = ds2['batsman'].unique()
count=0
def fours(x): 
    global count
    if x==4:
        count+=1

batsman_fours = []        
for var in batsman:
    temp = ds2[ds2['batsman']==var]
    temp['batsman_runs'].apply(fours)
    batsman_fours.append(count)
    count=0

Batsman=pd.DataFrame(batsman).reset_index()
Batsman.columns=['no','Player']
Fours=pd.DataFrame(batsman_fours).reset_index()
Fours.columns=['no','Fours']
Batsman=pd.merge(Batsman,Fours)
Batsman.drop('no',inplace=True,axis=1)
Batsman.sort_values('Fours', ascending=False, inplace=True)
Batsman=Batsman[:10]
Batsman

#PLOTTING

plt.figure(figsize=(11,7), facecolor='#B0E0E6')
plt.bar(x=Batsman['Player'], height=Batsman['Fours'])
plt.title('Top 10 Batsman with most number of FOURS')
plt.xlabel('BATSMAN')
plt.ylabel('FOURS')
plt.xticks(rotation=90)
plt.show()
```

## Q26). Top 10 Batsman with most number of SIXES.

```python
batsman = ds2['batsman'].unique()
count=0
def six(x): 
    global count
    if x==6:
        count+=1

batsman_six = []        
for var in batsman:
    temp = ds2[ds2['batsman']==var]
    temp['batsman_runs'].apply(six)
    batsman_six.append(count)
    count=0

Batsman=pd.DataFrame(batsman).reset_index()
Batsman.columns=['no','Player']
Six=pd.DataFrame(batsman_six).reset_index()
Six.columns=['no','Six']
Batsman=pd.merge(Batsman,Six)
Batsman.drop('no',inplace=True,axis=1)
Batsman.sort_values('Six', ascending=False, inplace=True)
Batsman=Batsman[:10]

#PLOTTING

plt.figure(figsize=(11,7), facecolor='#B0E0E6')
plt.bar(x=Batsman['Player'], height=Batsman['Six'])
plt.title('Top 10 Batsman with most number of SIXES')
plt.xlabel('BATSMAN')
plt.ylabel('SIXES')
plt.xticks(rotation=90)
plt.show()
```

## Q27). Top 10 Batsman with most number of dot balls

```python
batsman = ds2['batsman'].unique()
count=0
def six(x): 
    global count
    if x==0:
        count+=1

batsman_dot = []        
for var in batsman:
    temp = ds2[ds2['batsman']==var]
    temp['batsman_runs'].apply(six)
    batsman_dot.append(count)
    count=0

Batsman=pd.DataFrame(batsman).reset_index()
Batsman.columns=['no','Player']
Dot=pd.DataFrame(batsman_dot).reset_index()
Dot.columns=['no','Dot']
Batsman=pd.merge(Batsman,Dot)
Batsman.drop('no',inplace=True,axis=1)
Batsman.sort_values('Dot', ascending=False, inplace=True)
Batsman=Batsman[:10]

#PLOTTING

plt.figure(figsize=(11,7), facecolor='#B0E0E6')
plt.bar(x=Batsman['Player'], height=Batsman['Dot'])
plt.title('Top 10 Batsman with most number of DOTS')
plt.xlabel('BATSMAN')
plt.ylabel('DOTS')
plt.xticks(rotation=90)
plt.show()
```

## Q29). Orange Cap holder (Highest runs by Individual) in each Season

```python
season = ds1['season'].unique()

name,runs = [],[]
for var in season:
    df = ds1[ds1['season']==var]
    temp = ds2[ds2['match_id'].isin(df['id'].values)]
    temp = temp.groupby('batsman')['batsman_runs'].sum().reset_index()
    temp.sort_values('batsman_runs', inplace=True,ascending=False)
    temp = temp.iloc[0,:]
    name.append(temp['batsman'])
    runs.append(temp['batsman_runs'])
name=pd.DataFrame(name).reset_index()
name.columns=['no','Player']
runs=pd.DataFrame(runs).reset_index()
runs.columns=['no','Runs']
season=pd.DataFrame(season).reset_index()
season.columns=['no','Season']
run=pd.merge(season,name)
run=pd.merge(run,runs)
run.drop('no',inplace=True,axis=1)
run.sort_values('Season',inplace=True)
run
```

## Q30). Purple cap holder (highest wicket taker) in each Season

```python
season = ds1['season'].unique()

name,wickets = [],[]

#q = ds2['dismissal_kind'].unique()
out = ['caught', 'bowled', 'lbw', 'caught and bowled',  'stumped', 'hit wicket']

for var in season:
    df = ds1[ds1['season']==var]
    
    temp = ds2[ds2['match_id'].isin(df['id'].values)]
    temp = temp[temp['dismissal_kind'].isin(out)]
    temp = temp.groupby('bowler')['dismissal_kind'].count().reset_index()
    
    temp.sort_values('dismissal_kind', inplace=True,ascending=False)
    temp = temp.iloc[0,:]
   
    name.append(temp['bowler'])
    wickets.append(temp['dismissal_kind'])

purple = pd.DataFrame(data={'Season':season, 'Bowler':name, 'Wickets':wickets})
purple.sort_values('Season')
```
